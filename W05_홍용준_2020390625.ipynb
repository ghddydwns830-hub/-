```python
# Week05: 공급업체 평가 - EDA
# 홍용준 (2020390625)


# 1. 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# 2. 폰트 및 스타일 설정 (한글 깨짐 방지)
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
sns.set_style("whitegrid")

# 3. 데이터 불러오기
df = pd.read_csv("PurchasesFINAL12312016-1.csv")

# 4. 데이터 확인
print("데이터 크기:", df.shape)
print("컬럼 목록:", df.columns.tolist())
display(df.head(3))

# 5. 날짜형 컬럼 변환
date_cols = ['PODate', 'ReceivingDate', 'InvoiceDate', 'PayDate']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# 6. 새 변수 생성
df['LeadTime'] = (df['InvoiceDate'] - df['ReceivingDate']).dt.days
df['OrderTime'] = (df['ReceivingDate'] - df['PODate']).dt.days
df['PaymentLT'] = (df['PayDate'] - df['InvoiceDate']).dt.days

# 7. 결측치 제거
df = df.dropna(subset=['LeadTime', 'OrderTime', 'PaymentLT'])

# 8. 기초 통계
print("\n===== 기초 통계 요약 =====")
display(df[['LeadTime', 'OrderTime', 'PaymentLT']].describe().T)

# 9. 상관 분석 (Heatmap)
plt.figure(figsize=(8,6))
sns.heatmap(df[['LeadTime', 'OrderTime', 'PaymentLT']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# 10. 공급업체별 리드타임 평균
vendor_mean = df.groupby('VendorName')['LeadTime'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
vendor_mean.head(10).plot(kind='bar', color='steelblue')
plt.title('공급업체별 평균 리드타임 (상위 10개)')
plt.ylabel('리드타임 (일)')
plt.xticks(rotation=45)
plt.show()

# 11. 월별 리드타임 추이 (시계열)
df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_leadtime = df.groupby('Month')['LeadTime'].mean()

plt.figure(figsize=(10,5))
plt.plot(monthly_leadtime.index.astype(str), monthly_leadtime.values, marker='o', color='darkorange')
plt.title('월별 평균 리드타임 추이')
plt.xlabel('Month')
plt.ylabel('Average LeadTime (days)')
plt.grid(True)
plt.show()

# 12. 시계열 분해 (데이터가 충분한 경우에만 실행)
if len(monthly_leadtime) >= 24:
    result = seasonal_decompose(monthly_leadtime, model='additive', period=12)
    result.plot()
    plt.show()
else:
    print(f"\n⚠️ 데이터가 {len(monthly_leadtime)}개월치뿐이라 seasonal_decompose는 생략합니다.")
    print("→ 단순 추세 그래프로 대체했습니다.\n")

# 13. 요약 인사이트
print("===== 요약 인사이트 =====")
print(f"평균 리드타임: {df['LeadTime'].mean():.2f}일")
print("\n리드타임 상위 5개 공급업체:")
print(df.groupby('VendorName')['LeadTime'].mean().sort_values(ascending=False).head())

print("\n✅ 주요 관찰:")
print("- 리드타임과 결제 지연(PaymentLT)은 약한 음의 상관관계를 보임.")
print("- 일부 공급업체(VendorName)는 평균 리드타임이 길어 납기 효율성 개선 필요.")
print("- 월별 추세를 보면 특정 시점에 리드타임 급증 구간 존재 → 공급망 병목 가능성.")
print("- 향후 고리드타임 공급업체 중심으로 계약 조건 및 주문 분배 전략 검토 필요.")
```
