```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
sns.set(style="whitegrid")

df = pd.read_csv(r"C:\Users\ghddy\Projects\HelloPython\WA_Fn-UseC_-HR-Employee-Attrition.csv")


print("shape:", df.shape)
display(df.head(3))
display(df.describe(include="all").T)
print("missing_total:", int(df.isna().sum().sum()))
print("duplicate_rows:", int(df.duplicated().sum()))
print(df["Attrition"].value_counts())

num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in df.columns if c not in num_cols]

df["_Attrition_"] = (df["Attrition"] == "Yes").astype(int)

corr = df[num_cols + ["_Attrition_"]].corr()["_Attrition_"].sort_values(ascending=False)
top_corr = corr.drop("_Attrition_").head(12).index.tolist()

plt.figure(figsize=(8, 8))
sns.heatmap(
    df[top_corr + ["_Attrition_"]].corr(),
    annot=True, fmt=".2f", cmap="coolwarm", square=True
)
plt.title("ìƒìœ„ ìƒê´€ í”¼ì²˜ ìƒê´€ í–‰ë ¬")
plt.tight_layout()
plt.show()

plt.figure(figsize=(5, 4))
sns.countplot(x="Attrition", data=df)
plt.title("í‡´ì‚¬ ì—¬ë¶€ ë¶„í¬")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(data=df, x="Age", hue="Attrition", bins=25, multiple="stack")
plt.title("ë‚˜ì´ ë¶„í¬(í‡´ì‚¬ ì—¬ë¶€ë³„)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df)
plt.title("ì›”ê¸‰ vs í‡´ì‚¬ ì—¬ë¶€")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
sns.boxplot(x="Attrition", y="YearsAtCompany", data=df)
plt.title("ê·¼ì†ì—°ìˆ˜ vs í‡´ì‚¬ ì—¬ë¶€")
plt.tight_layout()
plt.show()

def rate_barplot(col, top_n=None, figsize=(7,4)):
    order = (
        df.groupby(col)["_Attrition_"]
        .mean()
        .sort_values(ascending=False)
    )
    if top_n:
        order = order.head(top_n)
    plt.figure(figsize=figsize)
    sns.barplot(x=order.index, y=order.values)
    plt.title(f"{col}ë³„ í‡´ì‚¬ìœ¨")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()

rate_barplot("Department")
rate_barplot("JobRole", figsize=(9,4))
rate_barplot("OverTime")
rate_barplot("BusinessTravel")
```
